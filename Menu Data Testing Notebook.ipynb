```python
locations_restaurants = dict()
for index, row in df_menus.iterrows():
    try:
        locations_restaurants[row['r_name']] = [(np.float64(row['r_lat']),np.float64(row['r_lon'])),]
    except:
        pass
```

```python
len(locations_restaurants)
```

```python
counter = 0
for r in locations_restaurants:
    try:
        avg_price = np.average((df_menus[df_menus['r_name']==r].item_price).apply(float))
        locations_restaurants[r] = [locations_restaurants[r][0], locations_restaurants[r][1], avg_price]
        print(avg_price)
    except:
        locations_restaurants[r] = [locations_restaurants[r], None]
    counter += 1
    if counter % 100 == 0:
        print(counter,"done")
```

```python
def inline_map(m, width=650, height=500):
    """Takes a folium instance and embed HTML."""
    m._build_map()
    srcdoc = m.HTML.replace('"', '&quot;')
    embed = HTML('<iframe srcdoc="{}" '
                 'style="width: {}px; height: {}px; '
                 'border: none"></iframe>'.format(srcdoc, width, height))
    return embed
```

```python
locations_restaurants['Cassava ']
```

```python
width, height = 750, 550
restaurants_loc = folium.Map(location=[37.759953, -122.44147], zoom_start=12,
                    tiles='OpenStreetMap', width=width, height=height)

for i in locations_restaurants:
#    color = [if locations_restaurants[i][1] == None: 10]
    restaurants_loc.circle_marker(location=(locations_restaurants[i][0][0][0],locations_restaurants[i][0][0][1]), popup=i, radius=10)

inline_map(restaurants_loc, width, height)
```
